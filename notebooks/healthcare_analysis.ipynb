### Q5 : Compare hospitals by revenue vs volume (which hospitals are high revenue but low volume and vice versa).

```python
hospital_metrics = health.groupby("hospital").agg(
    admissions=("hospital", "size"),
    total_billing=("bill", "sum"),
    avg_billing=("bill", "mean")
).reset_index()
```

```python
high_rev_low_vol = hospital_metrics[
    (hospital_metrics["total_billing"] >= hospital_metrics["total_billing"].quantile(0.90)) &
    (hospital_metrics["admissions"] <= hospital_metrics["admissions"].median())
].sort_values("total_billing", ascending=False)
```

```python
high_vol_low_rev = hospital_metrics[
    (hospital_metrics["admissions"] >= hospital_metrics["admissions"].quantile(0.90)) &
    (hospital_metrics["total_billing"] <= hospital_metrics["total_billing"].median())
].sort_values("admissions", ascending=False)
```

```python
high_vol_low_rev
```

### Q15 : Identify medical conditions that are low volume but high billing (high-value conditions).

```python
cond_metrics = health.groupby("medical_condition").agg(
    admissions=("medical_condition", "size"),
    total_billing=("bill", "sum"),
    avg_billing=("bill", "mean")
).reset_index()
```

```python
high_bill_low_vol = cond_metrics[
    (cond_metrics['avg_billing'] >= cond_metrics['avg_billing'].quantile(0.75)) &
    (cond_metrics['admissions'] <= cond_metrics['admissions'].quantile(0.25))
    ].sort_values('avg_billing', ascending=False)
high_bill_low_vol
```

### Q17 : Create Length of Stay (LOS) in days using Discharge Date minus Date of Admission.

### Q22 : Identify hospitals with unusually long LOS but low average billing (inefficiency risk)

```python
hosp_eff = health.groupby("hospital").agg(
    avg_los=("los", "mean"),
    avg_billing=("bill", "mean"),
    admissions=("hospital", "size")
).reset_index()
```

```python
high_los_low_bill = hosp_eff[
    (hosp_eff['avg_billing'] <= hosp_eff['avg_billing'].quantile(0.25)) &
    (hosp_eff['avg_los'] >= hosp_eff['avg_los'].quantile(0.75))
    ].sort_values('avg_los', ascending=False)
high_los_low_bill
```

### Q23 : Identify hospitals with short LOS but high average billing (efficient/high-value pattern).

```python
high_los_low_bill = hosp_eff[
    (hosp_eff['avg_billing'] >= hosp_eff['avg_billing'].quantile(0.25)) &
    (hosp_eff['avg_los'] <= hosp_eff['avg_los'].quantile(0.75))
    ].sort_values('avg_billing', ascending=False)
high_los_low_bill
```

### Q29 : Identify hospitals that depend heavily on a single insurance provider (concentration risk).

```python
hosp_ins = pd.crosstab(health['hospital'], health['insurer'])
```

```python
share = hosp_ins.div(hosp_ins.sum(axis=1), axis=0)
```

```python
share['max'] = share.max(axis=1)
```

```python
highly_dependent = share[share['max'] >= 0.70].sort_values('max', ascending=False)
highly_dependent.head(10)
```

### Q33 : Identify hospitals with unusually high emergency admission percentage.

```python
high_emergency = (pd.crosstab(health['hospital'], health['admission_type'], normalize='index') * 100)

emergency_col = None
for c in high_emergency.columns:
    if str(c).strip().lower() == "emergency":
        emergency_col = c
        break

if emergency_col is None:
    high_emergency
else:
    high_emergency[[emergency_col]].sort_values(emergency_col, ascending=False).head(15)
```
