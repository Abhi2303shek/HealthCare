```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
health = pd.read_csv('healthcare.csv', sep=',')
health
```

```python
health = health.rename(columns = {
    'Name' : 'name',
    'Age' : 'age',
    'Gender' : 'gender',
    'Blood Type' : 'blood_type',
    'Medical Condition' : 'medical_condition', 
    'Date of Admission' : 'doa', 
    'Doctor' : 'dr', 
    'Hospital' : 'hospital', 
    'Insurance Provider' : 'insurer', 
    'Billing Amount' : 'bill', 
    'Room Number' : 'room', 
    'Admission Type' : 'admission_type', 
    'Discharge Date' : 'discharge', 
    'Medication' : 'medicine', 
    'Test Results' : 'test'
})
```

```python
health['name'] = health['name'].str.title()
health
```

```python
health['doa'] = pd.to_datetime(health['doa'])
health['discharge'] = pd.to_datetime(health['discharge'])
```

```python
health['discharge'] = health['discharge'].dt.strftime('%d-%m-%Y')
```

```python
health
```

```python
health["month"] = health["doa"].dt.to_period("M").astype(str)
```

```python
health
```

## 🔹 SECTION 1: Revenue Architecture (Value Engine)


### 1. Plot total billing amount by hospital (Top 15).

```python
top15 = health.groupby('hospital')['bill'].sum().nlargest(15).reset_index()
```

```python
sns.barplot(data=top15, y='hospital', x='bill')
```

### 2. Plot hospital revenue vs volume
   (Scatter: X = admissions, Y = total billing).

```python
top10 = health.groupby('hospital').agg(admission=('hospital','count'), revenue=('bill','sum')).reset_index()
```

```python
sns.scatterplot(data=top10, x='admission', y='revenue')
```

### 3. Plot Pareto chart – Top hospitals contributing to 80% of total billing.


```python
pareto = health.groupby('hospital')['bill'].sum().sort_values(ascending=False)
```

```python
pareto_df = pareto.cumsum() / pareto.sum()
```

```python
sns.lineplot(x=range(len(pareto_df)), y=pareto_df.values)
```

### 4. Plot billing contribution (%) by admission type.


```python
bill_con = health.groupby('admission_type')['bill'].sum()
```

```python
bill_con_pct = (bill_con / bill_con.sum()).reset_index()
```

```python
sns.barplot(data=bill_con_pct, x='admission_type', y='bill')
```

### 5. Plot total billing by insurance provider.


```python
bill_ins = health.groupby('insurer')['bill'].sum().reset_index()
```

```python
sns.barplot(data=bill_ins, x='insurer', y='bill')
```

### 6. Plot top 10 doctors by total billing amount.


```python
top10 = health.groupby('dr')['bill'].sum().nlargest(10).reset_index()
```

```python
sns.barplot(data=top10, y='dr', x='bill')
```

## 🔹 SECTION 2: Operational Efficiency

### 7. Plot distribution of Length of Stay (LOS).


```python
sns.histplot(health['los'], kde=True)
```

### 8. Plot average LOS by hospital (Top 15 longest LOS).


```python
top15 = health.groupby('hospital')['los'].mean().nlargest(15).reset_index()
```

```python
sns.barplot(data=top15, x='los', y='hospital')
```

### 9. Plot average billing vs average LOS by hospital
   (Efficiency scatter map).

```python
efficiency = health.groupby('hospital').agg(avg_bill=('bill','mean'), avg_los=('los','mean')).reset_index()
```

```python
sns.scatterplot(data=efficiency, x='avg_bill', y='avg_los')
```

### 10. Plot LOS distribution by admission type.


```python
sns.boxplot(data=health, x='admission_type', y='los')
```

### 11. Plot monthly admissions trend.


```python
monthly_adm = health.groupby("month").size().reset_index(name="Admissions")
```

```python
sns.lineplot(data=monthly_adm, y='month', x='Admissions')
```
